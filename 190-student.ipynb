```python
#import the libraries


#Task 1
#Read the bmi.csv
import  numpy as np
import pandas as pd
from matplotlib import pyplot as plt


dataframe = pd.read_csv('bmi.csv')
df = dataframe.dropna()
df
```

```python
#Task 2
#Data is sorted in descending order in accordance with BMI value
#Find the top 5 age group where the BMI value is the highest, and plot a bar graph out of it
df = dataframe.head(5)
df
plt.xlabel("BMI")
plt.xticks(rotation='vertical')
plt.ylabel("Age")


label = df["BMI"] 
value = df["Age"]
plt.bar(label, value,width=0.8, color=('purple','yellow','green','blue' , 'red')) 
```

```python
#Task 5
#Read the insulin.csv
dataframe = pd.read_csv('insulin.csv')
df = dataframe.dropna()
df
```

```python
#Task 6
#Data is sorted in descending order in accordance with Insulin value
#Find out what will be the Glucose and BMI value when the Insulin is highest
df = dataframe.head(5)
df
value=df["Insulin"]
v=df["BMI"]
gul=df["Glucose"]
j=[]
q=[]
p=[]
for i in value:
    j.append(str(i))
for m in v:
    q.append(str(m))
for k in gul:
    p.append(str(k))
i=""
for i in range(0,5):
    print("Insulin - "+j[i]+" , BMI - "+q[i]+" , Glucose - "+p[i])
```
